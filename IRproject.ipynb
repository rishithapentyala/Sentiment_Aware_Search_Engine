```python
import pandas as pd
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from fuzzywuzzy import process
```

```python
# Initialize VADER
vader_analyzer = SentimentIntensityAnalyzer()
```

```python
def get_vader_sentiment(text):
    score = vader_analyzer.polarity_scores(str(text))['compound']
    if score >0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
def get_textblob_sentiment(text):
    score = TextBlob(str(text)).sentiment.polarity
    if score > 0.05:
        return 'Positive'
    elif score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis
df['vader_sentiment'] = df['Review'].apply(get_vader_sentiment)
df['textblob_sentiment'] = df['Review'].apply(get_textblob_sentiment)

# Display all unique product names
unique_products = df['product_name'].unique()
```

```python

import pandas as pd
import torch
from flair.models import TextClassifier
from flair.data import Sentence
from tqdm import tqdm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

classifier = TextClassifier.load('sentiment')

# Move to GPU if available
if torch.cuda.is_available():
    classifier.to('cuda')
    print("Using GPU for Flair inference.")
else:
    print("Using CPU for Flair inference.")

# Single prediction function with confidence thresholds
def predict_flair_sentiment_single(text, pos_conf=0.60, neg_conf=0.60):
    if pd.isna(text) or str(text).strip() == "":
        return "unknown"
    s = Sentence(str(text))
    classifier.predict(s)
    if not s.labels:
        return "unknown"
    label = s.labels[0]
    lab_value = label.value.lower()
    score = label.score
    if lab_value == 'positive' and score >= pos_conf:
        return 'positive'
    elif lab_value == 'negative' and score >= neg_conf:
        return 'negative'
    else:
        return 'neutral'

# Batch prediction
def predict_flair_batch(texts, batch_size=64, pos_conf=0.60, neg_conf=0.60):
    sentences = [Sentence(str(t)) if not pd.isna(t) else Sentence("") for t in texts]
    results = [None] * len(sentences)

    for i in tqdm(range(0, len(sentences), batch_size), desc="Flair batches"):
        batch = sentences[i:i+batch_size]
        classifier.predict(batch)
        for idx, s in enumerate(batch, start=i):
            if not s.labels:
                results[idx] = "unknown"
            else:
                label = s.labels[0]
                lab_value = label.value.lower()
                score = label.score
                if lab_value == 'positive' and score >= pos_conf:
                    results[idx] = 'positive'
                elif lab_value == 'negative' and score >= neg_conf:
                    results[idx] = 'negative'
                else:
                    results[idx] = 'neutral'
    return results


#  Run Flair on your reviews

pos_confidence_threshold = 0.60
neg_confidence_threshold = 0.60

df['flair_sentiment'] = predict_flair_batch(
    df['Review'].fillna("").tolist(),
    batch_size=64,
    pos_conf=pos_confidence_threshold,
    neg_conf=neg_confidence_threshold
)

# Quick peek
print(df[['Review','flair_sentiment']].head())

# Evaluate vs ground truth

true_sentiment = (
    df['Sentiment']
    .fillna("unknown")
    .astype(str)
    .str.lower()
    .str.strip()
)

pred_flair = (
    df['flair_sentiment']
    .fillna("unknown")
    .astype(str)
    .str.lower()
    .str.strip()
)

mask = (true_sentiment != "unknown") & (pred_flair != "unknown")
true_clean = true_sentiment[mask]
pred_clean = pred_flair[mask]

print("Flair Accuracy:", round(accuracy_score(true_clean, pred_clean) * 100, 2), "%")
print("\nFlair Classification Report:")
print(classification_report(true_clean, pred_clean))
print("\nFlair Confusion Matrix:")
print(confusion_matrix(true_clean, pred_clean))
```
